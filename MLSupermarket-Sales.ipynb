importacao das bibliotecas

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Teste de leitura de exemplos

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
display(df)
```

```python
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')
display(df)
```

limpeza de dados

```python
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)  # converter genero para binario
df['Customer type'] = df['Customer type'].replace({'Member': 0, 'Normal': 1})  # converter tipo de cliente para binario
display(df)
```

ajustar as features ...elas devem ser numericas

```python
# ajustando as features
# ajustando as features
features_simplified = df[['Unit price', 'Customer type']]  # Use double brackets to select multiple columns
targed = df['Gender']
display(features_simplified)
```

dividir o dataset

```python
# divisao do dataset
x_train, x_test, y_train, y_test = train_test_split(features_simplified, targed, test_size=0.2,random_state=42)
```

treinar o modelo

```python
# treinar o modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

acuria dos dados

```python
# acuria dos daddos
y_pred = model.predict(x_test)
```

calcular a acuria :calcular a predicao dos dados treinados

```python
#calculaar acuria
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {accuracy:.2f}')
```
